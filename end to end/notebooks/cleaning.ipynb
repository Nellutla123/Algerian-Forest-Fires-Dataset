DATACLEANING

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,'Region']=0
dataset.loc[122:,'Region']=1
df = dataset
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop= True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#setting spaces blw columns
df.columns = df.columns.str.strip()
df.columns
```

changing the required columns as int data type


```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
##changing the other columns to float data datatype
objects=[feature for feature in df.columns if df[feature].dtype == 'object']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
##lets save the cleaned dataset

df.to_csv('AlgerainCLEANED.csv',index=False)
```

EDA(Exploratery Data Analysis)

```python
#dropping day,year, month as they are not required
df_copy=df.drop(['day','month','year'],axis=1)
df_copy.head()
```

```python
#encoding of the categories in classe
print(df_copy['Classes'].unique())
df_copy['Classes'] = df_copy['Classes'].astype(str)
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire', case=False, na=False), 0, 1)

```

```python
print(df_copy['Classes'].value_counts())
print(df_copy.dtypes)  # Ensure it's now int
```
